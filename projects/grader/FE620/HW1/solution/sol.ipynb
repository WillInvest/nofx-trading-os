```python
import numpy as np
from scipy.optimize import fsolve

rate_schedule = [(1.0, 0.03), (2.0, 0.035), (5.0, 0.0425), (10.0, 0.045)]

def get_rate(t, schedule):
    return next(r for t_max, r in schedule if t <= t_max)

def price_bond(coupon, face, maturity, schedule):
    times = np.arange(0.5, maturity + 0.5, 0.5)
    cflows = [coupon/2] * (len(times)-1) + [coupon/2 + face]
    rates = [get_rate(t, schedule) for t in times]
    return sum(c * np.exp(-r * t) for c, r, t in zip(cflows, rates, times))

def flat_price(y, coupon, face, maturity):
    times = np.arange(0.5, maturity + 0.5, 0.5)
    cflows = [coupon/2] * (len(times)-1) + [coupon/2 + face]
    return sum(c * np.exp(-y * t) for c, t in zip(cflows, times))

B1 = price_bond(5.0, 100, 2, rate_schedule)
B2 = price_bond(6.0, 100, 10, rate_schedule)

y1 = fsolve(lambda r: flat_price(r, 5.0, 100, 2) - B1, 0.03)[0]
y2 = fsolve(lambda r: flat_price(r, 6.0, 100, 10) - B2, 0.04)[0]

print(f"B1 = {B1:.3f}, YTM = {y1*100:.3f}%")
print(f"B2 = {B2:.3f}, YTM = {y2*100:.3f}%")
```
